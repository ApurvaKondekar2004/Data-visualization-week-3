```python
#creating a special dataframe for gender
data=pd.DataFrame({'Ratio' : [len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])]},index=['Males','Females'])
data
```

THE NUMBER OF MALES AND FEMALES

```python
data.plot(kind='pie',y='Ratio',figsize=(6,6),autopct="%0.1f")

## So this pie chart is giving us the ratio of how many males and females shopped at black friday(the ratio of men is more as compared to women)
```

```python
df.groupby('Gender').size().plot(kind='bar')
```

HOW MUCH MALES HAVE SPENT IN COMPARISON WITH FEMALES

```python
df.groupby('Gender')['Purchase'].sum().plot(kind='pie',autopct="%0.1f")

## This graph tells males have purchased more as compared to females
```
